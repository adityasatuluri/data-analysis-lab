```python
import numpy as np
import pandas as pd
```

```python
#1.
iris = pd.read_csv('iris1.csv', delimiter=',', dtype=None, header=None)
species_dict = {'Iris-setosa': 1, 'Iris-versicolor': 2, 'Iris-virginica': 3}
iris[5] = iris[4].map(species_dict)  # iris[5]:result_column iris[4]:dependant_column
iris
```

```python
#2. Dropping column whose index=4 from array iris
iris=iris.drop([4],axis=1)
```

```python
#3.print shape, dimensions and size
print(iris.shape)
print(iris.size)
print(iris.ndim)
```

```python
#4. Splitting iris into 2d arrays
iris_arr=iris.to_numpy()
iris_arr = np.split(iris_arr, 3)
iris1=iris_arr[0]
iris2=iris_arr[1]
iris3=iris_arr[2]

#5. Print the arrays
print(iris1,'\n\n',iris2,'\n\n',iris3)
```
