3.	Write a pandas program to add, subtract, multiple and divide two pandas series.

```python
p1=pd.Series([30,31,32,33,34])
p2=pd.Series([2,3,4,5,6])
print("ADDITION")
print(p1+p2)
print("SUBTRACTION")
print(p1-p2)
print("MULTIPLICATION")
print(p1*p2)
print("DIVISION")
print(p1/p2)
```

4.	Write a Pandas program to convert all the string values to upper, lower cases in a given pandas series. Also find the length of the string values.

```python
p=pd.Series(['Adi','Harsha','bose','satu'])
print("LOWER CASE")
print(p.str.lower())
print("\nUPPER CASE")
print(p.str.upper())
print("\nLENGTHS")
print(p.str.len())
```

5.	Write a Pandas program to remove whitespaces, left sided whitespaces and right sided whitespaces of the string values of a given pandas series

```python
p=pd.Series(["  Adi","Harsha   ","bose   ","   satu"])
print("Given Series")
print(p)
print("\nRight Side Whitespaces Removal")
print(p.str.rstrip())
print("\nLeft Side Whitespaces Removal")
print(p.str.lstrip())
p=pd.Series(["  Adi","Harsha   ","bose   ","   satu"])
print("\nBoth Side Whitespaces Removal")
print(p.str.strip())
```

2. How to sort rows within a pandas dataframe

```python
import pandas as pd
d = {

    'Student Name': ['Aditya Satuluri', 'Luffy', 'Sanji', 'Ace'],
    'Father Name': ['SSRAO', 'Monkey D Dragon', 'Vinsmoke Judge', 'Gold D Rogers'],
    'Mobile Number': [123456789, 456123789, 123456789, 143254154],

}
index_labels=['Student 1','Student 2','Student 3','Student 4']
p=pd.DataFrame(d,index=index_labels)
print(p)
print("\n\t\tAfter Sorting\n")
p=p.sort_values(by='Student Name')
print(p)
```

7. Search pandas dataframe for a value

```python
p= pd.DataFrame([[10,20],[50,60]], columns=['a','b'], index=[1,2])
print(p)
value= int(input("Enter Value to search: "))
for i in p.columns:
  if (p[i]==value).any():
    print("Found in column {}".format(i))
    break
  else:
    print("Not Found")
```

13. How to rank a pandas dataframe in ascending and descending order


```python
d = {
    'Student Name': ['Aditya Satuluri', 'Luffy', 'Sanji', 'Ace'],
    'Father Name': ['SSRAO', 'Monkey D Dragon', 'Vinsmoke Judge', 'Gold D Rogers'],
    'Mobile Number': [123, 456, 789, 154],
    'RNo': [1, 2, 3, 4],
    'Grade':[8,4,7.8,4.8]
}
p=pd.DataFrame(d)
print(p)
print("\n\t\t\tRankings\n")
p['a_rank'] = p['Grade'].rank(method='average', ascending=True)
p['d_rank'] = p['Grade'].rank(method='average', ascending=False)
print(p)
```
